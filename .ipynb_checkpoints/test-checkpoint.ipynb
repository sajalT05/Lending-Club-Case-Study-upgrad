```python
import pandas as pd
```

```python
df=pd.read_csv('loan.csv')
```

```python
df
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
lower_quantile=Q1 - 1.5 * IQR
upper_quantile=Q3 + 1.5 * IQR
df = df[~((df < lower_quantile)|(df > upper_quantile).any(axis=1))]
```

```python
df.shape
```

```python
df['annual_inc'].plot(kind='box')
```
