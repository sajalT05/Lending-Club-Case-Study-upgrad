# meaning of different columns after google search
Most of the column meaning are taken directly from https://www.cs.dartmouth.edu/~lorenzo/teaching/cs174/Archive/Winter2015/Projects/proposals/fmz.pdf. SOme are from https://rstudio-pubs-static.s3.amazonaws.com/303215_bb3e159757314931ac706574b96162b7.html and https://www.rubydoc.info/gems/lending_club/0.0.2/LendingClub/Loan
- id: id assigned to the loan after approval
- member_id: A unique Lending Club(LC) assigned Id for the borrower member.
- loan_amnt: loan amount applied for
- funded_amnt: The total amount committed to that loan at that point in time.
- funded_amnt_inv: The total amount committed by investors for that loan at that point in time.
- term: the loan tenure
- int_rate: the interest rate on loan
- installment: the EMI amount
- grade: Lending club assigned loan grade
- sub_grade: LC assigned loan subgrade
- emp_title: Borrower job title at the time of loan application
- emp_length: employment length of borrower
- home_ownership: does borrower own house
- annual_inc: annual income of borrower at the time of application
- verification_status: if the income was verified by LC
- issue_d: loan issue month
- loan_status: current status of loan
- pymnt_plan: if a paymne tplan exists for the loan
- url: URL for the LC page with listing data
- desc: Loan description provided by the borrower
- purpose: purpose for taking the loan
- title: the loan title provided by the borrower
- zip_code: the zip code of the borrower
- addr_state: state of borrower address
- dti: total monthly debt payments divided by borrower's monthly income
- delinq_2yrs: the number of 30+ days past due incidences of delinquency in borrower credit file in last 2 yrs
- earliest_cr_line: month of the borrower's earliest credit line opning
- inq_last_6mths: number of inquiries by creditors in last 6 months
- mths_since_last_delinq: number of months since last delayed payment(delinquency)
- mths_since_last_record: number of months since the last public record
- open_acc: the number of open credit lines in borrower's credit file
- pub_rec: number of derogatory public records like bankruptcy etc
- revol_bal: Total credit revolving balance
- revol_util: amount of credit borrower is using relative to all available revolving credit
- total_acc: totla number of credit lines currently in the borrower's credit file
- initial_list_status: initial listing status of loan (W-WHole loans; can be purchased by investors fully, F-fractional loans)
- out_prncp: Remaining outstanding principal for total amount funded
- out_prncp_inv: Remaining outstanding principal for amount funded by investors
- total_pymnt: Payments recieved till date for the loan for total amount funded
- total_pymnt_inv: Payments recieved till date for the loan for amount funded by investors
- total_rec_prncp: Principal recieved till date
- total_rec_int: interest recieved till date
- total_rec_late_fee: late fees recieved till date
- recoveries: post charge off gross recovery
- collection_recovery_fee: post charge off collection fee; incurred to collect the remaining amount
- last_pymnt_d: last month payment was recieved
- last_pymnt_amnt: last total payment amount recieved
- next_pymnt_d: next scheduled payment date
- last_credit_pull_d: The most recent month LC pulled credit for this loan; assume after loan approval
- collections_12_mths_ex_med: Number of collections in last 12 months excluding medical collections
- policy_code: publicly available policy code 1, new products not publicly available code 2
- application_type: Indicates whether the loan is an individual application or a joint application with two co-borrowers
- acc_now_delinq: number of accounts on which borrower is now delinquent
- chargeoff_within_12_mths: NOT SURE
- delinq_amnt: past due amount owed for the accounts for the accounts on which the borrower is now delinquent
- pub_rec_bankruptcies: Number of public recorded bankruptcies
- tax_liens: number of tax liens(order for tax authorities to collect tax by siezing assets)

## remove outliers

```python
def is_not_outlier(data):
    q1 = np.quantile(data, 0.25)
    q3 = np.quantile(data, 0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return (lower_bound <= data) & (data <= upper_bound)
```
